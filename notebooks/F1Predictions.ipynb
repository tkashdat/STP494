```python
circuits = pd.read_csv("/Users/tk/ASU/STP494/data/circuits.csv")
circuits.head()
```

```python
circuits.drop(columns=['url', 'country', 'location', 'name', 'circuitRef'], inplace=True)
```

```python
circuits.info()
```

```python
racesdf = pd.merge(races, circuits, how='outer', on='circuitId')
```

```python
racesdf.sort_values(by=['raceId'], inplace=True)
racesdf.head()
```

```python
racesdf.info()
```

```python
racesdf = racesdf.dropna()
```

```python
racesdf = racesdf.astype({"raceId": int, "year": int, "round": int})
racesdf.head()
```

```python
racesdf = racesdf.loc[~racesdf['time'].str.contains('N')]
```

```python
pd.to_datetime(racesdf['date'], format='%d/%m/%y')
```

```python
racesdf.reset_index(inplace=True)
```

```python
racesdf.drop('index', axis=1, inplace=True)
```

```python
for i in range(len(racesdf)):
    if len(racesdf.loc[i,'time']) < 8:
        racesdf.loc[i,'time'] = '0' + racesdf.loc[i,'time']
        i += 1
    else:
        i += 1
```

```python
for i in range(len(racesdf)):
    racesdf.loc[i,'date'] = racesdf.loc[i,'date'].strftime("%Y-%m-%d")
    i += 1
```
